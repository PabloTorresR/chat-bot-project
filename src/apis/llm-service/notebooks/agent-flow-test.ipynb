```python
from core.llms.langchain.config import OPEN_AI_CONFIG, OPEN_AI_LIGHT_CONFIG
from langchain_openai import ChatOpenAI
from core.llms.langchain.chat_prompts.prompt_factory import CustomChatPromptFactory
from core.llms.langchain.templates.human.human_template import (
    human_template,
)
from core.llms.langchain.templates.human.human_template_with_history import (
    human_template_with_history,
)
from core.llms.langchain.multi_agents.agents.parameter_extractor import (
    ParameterExtractorAgent,
)
from core.llms.langchain.multi_agents.agents.word_generator import WordGeneratorAgent
from core.llms.langchain.multi_agents.agents import example_generator


from typing import List
from typing_extensions import TypedDict, Annotated
from langgraph.graph import MessagesState, START, END
from langchain_core.messages.human import HumanMessage
from langchain_core.messages.system import SystemMessage


llm = ChatOpenAI(
    model=OPEN_AI_CONFIG["model"],
    temperature=OPEN_AI_CONFIG["temperature"],
    api_key=OPEN_AI_CONFIG["api_key"],
)

llm_light = ChatOpenAI(
    model=OPEN_AI_LIGHT_CONFIG["model"],
    temperature=OPEN_AI_LIGHT_CONFIG["temperature"],
    api_key=OPEN_AI_LIGHT_CONFIG["api_key"],
)
```

```python
data_extractor = ParameterExtractorAgent(llm=llm_light)
result = await data_extractor.extract(
    prompt="foootballlllllll",
    messages=[
        HumanMessage("Hello I want to learn"),
        SystemMessage(
            "Hi, I can help you with that, what topics are you interested in?"
        ),
        HumanMessage("Words abot football"),
    ],
)
print(result)
```

```python
word_generator_agent = WordGeneratorAgent(llm=llm_light)
result = await word_generator_agent.generate(
    topics=["football"],
    language="spanish",
    native_language="english",
)
print(result)
```

```python
example_generator_agent = example_generator.ExamplesGeneratorAgent(llm=llm)
result = await example_generator_agent.generate(
    words=["persona", "stupida", "comida"],
    language="spanish",
    native_language="english",
)

print(result)
import json

formatted_json = json.dumps(result, indent=4)
print(formatted_json)
```

```python
from core.llms.langchain.multi_agents.agents.response import ResponseAgent


respnse_agent = ResponseAgent(llm=llm)
result = await respnse_agent.generate(
    words=[],
    topics=["football"],
    language="spanish",
    native_language="english",
    prompt="foootballlllllll",
    messages=[
        HumanMessage("Hello I want to learn"),
        SystemMessage(
            "Hi, I can help you with that, what topics are you interested in?"
        ),
        HumanMessage("Words abot football"),
    ],
)
print(result)
```
