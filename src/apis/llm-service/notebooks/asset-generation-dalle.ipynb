# Generate assets using DALL-E


```python
from openai import OpenAI

client = OpenAI()

app_style = """In style that is vibrant, flat, vector-art. Use a bold, bright color palette with teal, purple. The overall aesthetic is cyberpunk, with clean lines, simple and sharp geometric shapes, and minimal shading. Ensure the lines are sharp and clean, no round edges, with minimal shading."""

response = client.images.generate(
    model="dall-e-3",
    prompt=app_style
    + " a female anthropomorphic komodo dragon. ONLY BUST. She is friendly and smiling. Transparent background. Flat colors and design in 2D. Use only teal and purple",
    size="1024x1024",
    quality="hd",
    n=1,
)
```

```python
print(response.data[0].url)
```
