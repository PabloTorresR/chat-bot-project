# Generate assets using DALL-E


```python
from openai import OpenAI

client = OpenAI()

app_style = """In style that is vibrant, flat, vector-art. Use a bold, bright color palette with teal, purple, yellow and pink. The overall aesthetic is cyberpunk, with clean lines, simple and sharp geometric shapes, and minimal shading. Ensure the lines are sharp and clean, no round edges, with minimal shading."""

response = client.images.generate(
    model="dall-e-3",
    prompt=app_style
    + " a female komodo dragon in cyberpunk outfit crossing arms. It's raining phones with sad emoji in the screen. The background is a cyberpunk cityscape with neon lights and a futuristic aesthetic.",
    size="1024x1792",
    quality="hd",
    n=1,
)
```

```python
print(response.data[0].url)
```
